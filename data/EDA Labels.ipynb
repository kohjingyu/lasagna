```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
label_counts = np.load("label_count.npy")
nonzero_idx = np.nonzero(label_counts)
```

```python
plt.figure(figsize=(10,5))
plt.ylabel("Number of Occurrences")
plt.xlabel('Ingredient ID')
plt.title("Distribution of Ingredients")
plt.plot(label_counts[nonzero_idx], c="salmon")
```
