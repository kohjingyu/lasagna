```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
label_counts = np.load("label_count.npy")
label_counts[1] = 1
nonzero_idx = np.nonzero(label_counts)
```

```python
nonzero_idx = np.nonzero(label_counts)[0]

current_mapping = {}
for idx in nonzero_idx:
    current_mapping[idx] = len(current_mapping)
```

```python
plt.figure(figsize=(10,5))
plt.ylabel("Number of Occurrences")
plt.xlabel('Ingredient ID')
plt.title("Distribution of Ingredients")
plt.plot(label_counts[nonzero_idx][2:], c="salmon")
```
